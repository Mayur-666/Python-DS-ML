```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.displot(tips["total_bill"], color="r", kde=True)
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips)
```

```python
sns.pairplot(tips, hue="sex", palette ="coolwarm")
```

```python
sns.rugplot(tips["total_bill"])
```

```python
sns.barplot(x="sex",y="total_bill",data=tips, hue="sex", estimator=np.std)
```

```python
sns.countplot(x="sex", data=tips)
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue="smoker")
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
mod_tips = tips.drop(["sex","smoker","day","time"], axis=1)
mod_tips.head()
```

```python
tips_correlation = mod_tips.corr()
```

```python
# correlation heatmap
sns.heatmap(tips_correlation,cmap = 'viridis', vmin = -1, vmax = 1, center = 0, annot=True, fmt=".2f", square=True, linecolor="black", linewidths=1)
```

```python
fp = flights.pivot_table(index = "month", columns = "year", values= "passengers", observed=False)
fp.head()
```

```python
sns.heatmap(fp, linecolor="white", linewidths=1)
```

```python
import scipy
sns.clustermap(fp, standard_scale=1)
```
