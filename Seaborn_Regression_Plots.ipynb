# Regression Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex", markers=["+", "x"])
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, col="sex", hue="sex")
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, col="sex", row="time", hue="day")
```
