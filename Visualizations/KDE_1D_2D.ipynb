# 1D and 2D KDE

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# random data
data = np.random.randn(1000)
```

```python
# Plot for 1D KDE
sns.kdeplot(data, fill=True, color="blue")
plt.title("1D KDE Plot")
plt.show()
```

```python
# random data for two variables
x = np.random.randn(1000)
y = np.random.randn(1000)
```

```python
# Plot for 2D KDE
sns.kdeplot(x=x, y=y, cmap="Blues", fill=True)
plt.title("2D KDE Plot")
plt.show()
```
