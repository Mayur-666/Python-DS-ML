```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
x
```

```python
y = x**2
y
```

```python
plt.plot(x, y, "r-", label="Square Function")
plt.xlabel("X Lanel")
plt.ylabel("Y Label")
plt.title("Square")
plt.legend()
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, "b", label="Square Function")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(y, x, "y", label="Reverse Square Function")
plt.legend()

plt.show()
```

```python
# using oops way
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.6, 0.3, 0.2])

axes1.plot(x, y, "r")
axes1.set_title("Large Plot")

axes2.plot(y, x, "y")
axes2.set_title("Small Plot")

plt.show()
```

```python
plt.plot(x,x, label="Linear", marker="+")
plt.plot(x, y, label="Square", marker="o")
plt.plot(x, x-x, label="Constant", marker=".")
plt.legend()
plt.show()
```
