# Seaborn Grids

```python
import seaborn as sns
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris)
```

```python
import matplotlib.pyplot as plt
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g_new = sns.PairGrid(iris)
g_new.map_diag(sns.histplot) # used histplot inplace of distplot as it is deprecated
g_new.map_upper(plt.scatter)
g_new.map_lower(sns.kdeplot)
```

## Facet Grids

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
ge = sns.FacetGrid(data = tips, col = "time",row = "smoker")
ge.map(sns.histplot, 'total_bill')
```

```python
ge_new = sns.FacetGrid(data = tips, col = "time",row = "smoker")
ge_new.map(plt.scatter, 'total_bill', 'tip')
```
