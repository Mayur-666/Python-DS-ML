```python
# pandas concat
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "A" : ["A0", "A1", "A2", "A3"],
    "B" : ["B0", "B1", "B2", "B3"],
    "C" : ["C0", "C1", "C2", "C3"],
    "D" : ["D0", "D1", "D2", "D3"]},
    index= [0, 1, 2, 3])

df2 = pd.DataFrame({
    "A" : ["A4", "A5", "A6", "A7"],
    "B" : ["B4", "B5", "B6", "B7"],
    "C" : ["C4", "C5", "C6", "C7"],
    "D" : ["D4", "D5", "D6", "D7"]},
    index= [4, 5, 6, 7])

df3 = pd.DataFrame({
    "A" : ["A8", "A9"],
    "B" : ["B8", "B9"],
    "C" : ["C8", "C9"],
    "D" : ["D8", "D9"]},
    index= [8, 9])
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
pd.concat([df1, df2, df3])
```

```python
pd.concat([df1, df2, df3], axis = 1)
```
